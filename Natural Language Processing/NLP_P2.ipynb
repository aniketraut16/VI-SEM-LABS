## Displaying Statistics of Diffrent Corpuses

```python
def print_corpus_stats(corpus, corpus_name):
    words = corpus.words()
    sentences = corpus.sents() if hasattr(corpus, 'sents') else None
    print(f"Corpus: {corpus_name}")
    print(f"Tagged: {'Yes' if hasattr(corpus, 'tagged_words') else 'No'}")
    print(f"Total Words: {len(words)}")
    if sentences:
        print(f"Total Sentences: {len(sentences)}")
    print("-" * 50)

print_corpus_stats(gutenberg, "Gutenberg")
print_corpus_stats(reuters, "Reuters")
print_corpus_stats(webtext, "Webtext")
print_corpus_stats(genesis, "Genesis")
print_corpus_stats(brown, "Brown")
```

```python
# Categories in Brown Corpus
brown.categories()
```

```python
corpus = ' '.join(brown.words(categories='fiction'))
```

## Word and Sentense Tokenization

```python
words_token = word_tokenize(corpus)
sentence_token = sent_tokenize(corpus)

print(f'Words in Corpus:{len(words_token)}')
print(f'Sentence in Corpus:{len(sentence_token)}')
```

## Display unique words

```python
unique_words = set(words_token)
print("\nUnique words:", unique_words)
print(f"Number of unique words: {len(unique_words)}")
```

## Remove punctuation

```python
translator = str.maketrans('', '', string.punctuation)
words_token = [word.translate(translator) for word in words_token]
words_token = [word for word in words_token if word]
print(f'Words in Corpus after removing Puntuations:{len(words_token)}')
```

## Stop word removing

```python
print(f'Before Removing Stop Words:{len(words_token)}')
stop_words = set(stopwords.words('english'))
filtered_words = [w for w in words_token if not w.lower() in stop_words]
print(f'Before Removing Stop Words:{len(filtered_words)}')
print("Filtered words:", filtered_words)
```

## Stemming

```python
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_words]
```


## Part-of-Speech tagging

```python
pos_tags = nltk.pos_tag(filtered_words)

print("Stemmed words:", stemmed_words)
print("POS tags:", pos_tags)
```
