```python
image5 = cv2.imread('/content/drive/MyDrive/DIVP LAB/cameraman.jpg',0)
print("Original Picture")
cv2_imshow(image5)
print(image5.shape)
```

```python
mul1 = cv2.multiply(image5,.5)
print("Multiplied by 0.5")
cv2_imshow(mul1)
print("\n")

mul2 = cv2.multiply(image5,1)
print("Multiplied by 1")
cv2_imshow(mul2)
print("\n")

mul3 = cv2.multiply(image5,1.5)
print("Multiplied by 1.5")
cv2_imshow(mul3)
print("\n")

mul4 = cv2.multiply(image5,2)
print("Multiplied by 2")
cv2_imshow(mul4)
```

```python
div1 = cv2.divide(image5,2)
print("Divided by 2")
cv2_imshow(div1)
print("\n")

div2 = cv2.divide(image5,4)
print("Divided by 4")
cv2_imshow(div2)
print("\n")

div3 = cv2.divide(image5,6)
print("Divided by 6")
cv2_imshow(div3)
print("\n")

div4 = cv2.divide(image5,8)
print("Divided by 8")
cv2_imshow(div4)
```

```python
image6 = cv2.imread('/content/drive/MyDrive/DIVP LAB/Logical1.jpg',0)
cv2_imshow(image6)
print(image6.shape)
```

```python
image7 = cv2.imread('/content/drive/MyDrive/DIVP LAB/logical2.jpg',0)
cv2_imshow(image7)
print(image7.shape)
```

```python
image6 = cv2.resize(image6, (155, 152))
cv2_imshow(image6)
print(image6.shape)
```

```python
AND = cv2.bitwise_and(image6,image7)
cv2_imshow(AND)
```

```python
OR = cv2.bitwise_or(image6,image7)
cv2_imshow(OR)
```

```python
XOR = cv2.bitwise_xor(image6,image7)
cv2_imshow(XOR)
```

```python
image8 = cv2.imread('/content/drive/MyDrive/DIVP LAB/NOT operation.jpg',0)
cv2_imshow(image8)
print(image8.shape)
```

```python
NOT = cv2.bitwise_not(image8)
cv2_imshow(NOT)
```
